The `preprocess_text` function preprocesses text data by performing tokenization, stop word removal, and lemmatization. It takes an input text string, tokenizes it into individual words, removes common English stop words, lemmatizes the remaining tokens to their base or dictionary form, and then joins the tokens back into a single string.

```python
# create preprocess_text function
def preprocess_text(text):

    # Tokenize the text
    tokens = word_tokenize(text.lower())

    # Remove stop words
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]


    # Lemmatize the tokens
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

    # Join the tokens back into a string
    processed_text = ' '.join(lemmatized_tokens)
    return processed_text

start_time = dt.datetime.now()
print("Started text preprocessing at ", start_time)
df['Review'] = df['Review'].apply(preprocess_text)
end_time = dt.datetime.now()
elapsed_time = end_time - start_time
print("Finished text preprocessing at ", end_time)
print ("Text preprocessing completed. Elapsed time: ", elapsed_time)
```

For columns with integer values we fill the empty values with the mean of the non-empty values (imputation) and wherever the values are strings we have replaced the values with **"Unknown"**

```python
# Impute null values in 'Aircraft' column with 'Unknown'
df['Aircraft'].fillna('Unknown', inplace=True)

# Impute null values in 'Type Of Traveller' column with 'Unknown'
df['Type Of Traveller'].fillna('Unknown', inplace=True)

# Impute null values in 'Seat Type' column with 'Unknown'
df['Seat Type'].fillna('Unknown', inplace=True)

# Impute null values in 'Route' column with 'Unknown'
df['Route'].fillna('Unknown', inplace=True)

# Impute null values in 'Date Flown' column with 'Unknown'
df['Date Flown'].fillna('Unknown', inplace=True)

# Impute null values in 'Seat Comfort' column with the mean of non-null values
mean_seat_comfort = df['Seat Comfort'].mean()
df['Seat Comfort'].fillna(int(mean_seat_comfort), inplace=True)

css = df['Cabin Staff Service'].mean()
df['Cabin Staff Service'].fillna(int(css), inplace=True)

fab = df['Food & Beverages'].mean()
df['Food & Beverages'].fillna(int(fab), inplace=True)

gs = df['Ground Service'].mean()
df['Ground Service'].fillna(int(gs), inplace=True)

# Impute null values in 'Value For Money' column with the mean of non-null values
mean_value_for_money = df['Value For Money'].mean()
df['Value For Money'].fillna(mean_value_for_money, inplace=True)

# Impute null values in 'Inflight Entertainment' column with 'Not Available'
df['Inflight Entertainment'].fillna('Not Available', inplace=True)

# Impute null values in 'Wifi & Connectivity' column with 'Not Available'
df['Wifi & Connectivity'].fillna('Not Available', inplace=True)

# You can continue imputing null values for other columns as needed

# Display the first few rows of the DataFrame to verify changes
print(df.head())
```

**Feature Engineering**

Adding a new feature called `Sentiment` which contains values `Positive` or `Negative` to further enhance model's knowledge on happiness level of the customer on the specific airline

```python
df['Sentiment'] = df['Recommended'].apply(lambda x: 'Positive' if x == 'yes' else 'Negative')
```

Here we add another column called `Location Entities` which contains arrays of location values extracted from `Reviews` using **Named Entity Recognition(NER)** techinques

```python
import spacy
```

```python
# Load spaCy model for English
nlp = spacy.load("en_core_web_sm")
data=df['Review']
def extract_location_entities(review):
    doc = nlp(review)
    locations = [ent.text for ent in doc.ents if ent.label_ == "GPE"]  # GPE: geopolitical entity
    return locations
# Apply NER to each review and store the extracted entities in a new column
df["Location Entities"] = df["Review"].apply(extract_location_entities)
```

Also I created a column `Travel Day` to indicate the day on which they travelled to make the model understand if there is a relation between travelling on Weekdays or on Weekends indicated by column `Weekend Travel`

```python
# Replace invalid entries in 'Date Flown' column with NaN
df['Date Flown'] = df['Date Flown'].replace('Unknown', pd.NaT)

# Convert 'Date Flown' column to datetime with a specific format, ignoring NaN values
df['Date Flown'] = pd.to_datetime(df['Date Flown'], format='%B %Y', errors='coerce')

# Create a new column 'Travel Day' to extract the day of the week
df['Travel Day'] = df['Date Flown'].dt.day_name()

# Create a new column 'Weekend Travel' to indicate whether the travel day is a weekend or not
df['Weekend Travel'] = df['Travel Day'].apply(lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday')
```

Then I made another column `Review Length to Rating Ratio`  to measure the ratio of **Review Length to Rating ratio** to knw how ratings affected the review length. The more happier customers generally tend to add more to the reviews suugesting a more happy score.

```python
df['Overall_Rating'] = pd.to_numeric(df['Overall_Rating'], errors='coerce')

# Calculate the length of each review
df['Review Length'] = df['Review'].apply(len)

# Create a new column for the Review Length to Rating Ratio
df['Review Length to Rating Ratio'] = df['Review Length'] / df['Overall_Rating']
```
